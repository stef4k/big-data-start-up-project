```python
import firebase_admin
from firebase_admin import credentials, auth, exceptions
from firebase_admin.auth import UserImportHash
import pandas as pd
import hmac
import hashlib

if not firebase_admin._apps:
    cred = credentials.Certificate('firebase-auth.json')
    firebase_admin.initialize_app(cred)
```

```python
from firebase_admin import auth
# Start listing users from the beginning, 1000 at a time.
page = auth.list_users()
users=[]
while page:
    for user in page.users:
        users.append(user.uid)
    # Get next batch of users.
    page = page.get_next_page()

result = auth.delete_users(users)

print('Successfully deleted {0} users'.format(result.success_count))
print('Failed to delete {0} users'.format(result.failure_count))
for err in result.errors:
    print('error #{0}, reason: {1}'.format(result.index, result.reason))
```

```python
csv_file_path = 'csv/auth_users.csv' 
df = pd.read_csv(csv_file_path)
hash_alg = UserImportHash.hmac_sha256(key=b'secret_key')  # Replace 'secret_key' with your actual secret key

# Function to hash the password using HMAC-SHA256
def hash_password(password, salt):
    return hmac.new(salt, password.encode('utf-8'), hashlib.sha256).digest()

users = []

# Process each row in the CSV
for _, row in df.iterrows():
    email = row['email']
    password = row['password']  
    uid = row["id"]
    salt = b'some_random_salt_value'  

    # Hash the password
    password_hash = hash_password(str(password), salt)
    # Create ImportUserRecord for each user
    user_record = auth.ImportUserRecord(
        uid=str(uid),
        email=email,
        password_hash=password_hash,
        password_salt=salt
    )
    users.append(user_record)

# Import users
try:
    result = auth.import_users(users, hash_alg=hash_alg)
    print('Successfully imported {0} users. Failed to import {1} users.'.format(
        result.success_count, result.failure_count))
    for err in result.errors:
        print('Failed to import {0} due to {1}'.format(users[err.index].uid, err.reason))
except exceptions.FirebaseError as e:
    print(f"An error occurred: {e}")
```

```python
from firebase_admin import firestore
import csv

store = firestore.client()

file_path = "csv/private_users.csv"
collection_name = "private_users"


def batch_data(iterable, n=1):
    l = len(iterable)
    for ndx in range(0, l, n):
        yield iterable[ndx:min(ndx + n, l)]


data = []
headers = []
with open(file_path) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            for header in row:
                headers.append(header)
            line_count += 1
        else:
            obj = {}
            for idx, item in enumerate(row):
                obj[headers[idx]] = item
            data.append(obj)
            line_count += 1
    print(f'Processed {line_count} lines.')

for batched_data in batch_data(data, 499):
    batch = store.batch()
    for data_item in batched_data:
        doc_ref = store.collection(collection_name).document(data_item["id"])
        data_item.pop("id")
        batch.set(doc_ref, data_item)
    batch.commit()

print('Done')

file_path = "csv/public_users.csv"
collection_name = "public_users"
data = []
headers = []
with open(file_path) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            for header in row:
                headers.append(header)
            line_count += 1
        else:
            obj = {}
            for idx, item in enumerate(row):
                obj[headers[idx]] = str(item)
            data.append(obj)
            line_count += 1
    print(f'Processed {line_count} lines.')

for batched_data in batch_data(data, 499):
    batch = store.batch()
    for data_item in batched_data:
        doc_ref = store.collection(collection_name).document(data_item["username"])
        data_item.pop("username")
        batch.set(doc_ref, data_item)
    batch.commit()
print('Done')

file_path = "csv/tech_dummy.csv"
collection_name = "tech_dummy"
data = []
headers = []
with open(file_path) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            for header in row:
                headers.append(header)
            line_count += 1
        else:
            obj = {}
            for idx, item in enumerate(row):
                obj[headers[idx]] = str(item)
            data.append(obj)
            line_count += 1
    print(f'Processed {line_count} lines.')

for batched_data in batch_data(data, 499):
    batch = store.batch()
    for data_item in batched_data:
        doc_ref = store.collection(collection_name).document(data_item["id"])
        data_item.pop("id")
        batch.set(doc_ref, data_item)
    batch.commit()
print('Done')
```
